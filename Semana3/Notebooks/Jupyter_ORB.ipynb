```python
import cv2 #Librería Opencv
from matplotlib import pyplot as plt #Librería para mostrar imágenes

img = cv2.imread('/home/dani/Descargas/1.JPG',0) #Lee la imagen

orb = cv2.ORB_create(4000) #Inicia el detector; Threshold=4000
kp = orb.detect(img,None) #Detecta los keypoints
kp, des = orb.compute(img, kp) #Computa los descriptores
img2 = cv2.drawKeypoints(img,kp,None,color=(255,255,0), flags=0) #Dibuja los keypoints en la imagen
plt.imshow(img2),plt.show() #Muestra la imagen con los keypoints dibujados
print len(kp) #Muestra la cantidad de keypoints
```
